```python
dict1 ={'Customer_id':[1,2,3,4,5,6],'Product':['Oven','Oven','Oven','Oven','Oven','Oven']}
df1=pd.DataFrame(dict1)
df1
```

```python
inner_join_df=pd.merge(df1,df2,on='Customer_id',how='inner')
inner_join_df
```

```python
outer_join_df=pd.merge(df1,df2,on='Customer_id',how='outer')
outer_join_df
```

```python
import pandas as pd

# Sample dataframes
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
})

# Inner Join
inner_join = pd.merge(df1, df2, how='inner', on='ID')

# Outer Join
outer_join = pd.merge(df1, df2, how='outer', on='ID')

# Left Join
left_join = pd.merge(df1, df2, how='left', on='ID')

# Right Join
right_join = pd.merge(df1, df2, how='right', on='ID')

# Displaying results
print("Inner Join:\n", inner_join)
print("Outer Join:\n", outer_join)
print("Left Join:\n", left_join)
print("Right Join:\n", right_join)
```
